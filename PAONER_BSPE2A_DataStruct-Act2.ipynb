```python
#(Z) 1st matrix (initial)
Z = [[10, 4, 16],
     [15, 5, 1]]

#(Y) 2nd matrix (student number)
Y = [[2, 0, 2],
     [1, 3, 2]]

result = [[0, 0, 0],
          [0, 0, 0]]

print("1st Matrix:")
for row in Z:
    print(row)

print("2nd Matrix:")
for row in Y:
    print(row)

#(X) 3rd matrix (addition)
for i in range(len(Z)):
    for j in range(len(Z[0])):
        result[i][j] = Z[i][j] + Y[i][j]

print("3rd Matrix:")
for row in result:
    print(row)

#(W) 4th matrix (scalar multiplication)
def scalar(k,Z):
    m=len(Z)
    n=len(Z[0])
    return[[k*i for i in row]for row in Z]

W = scalar(2, Z)
print("4th Matrix:")
for row in W:
    print(row)

#(V) 5th matrix (transpose)
def transpose(Y):
    rows = len(Y)
    cols = len(Y[0])
    result = [[0 for b in range(rows)]for c in range(cols)]

    for b in range(rows):
        for c in range(cols):
            result[c][b] = Y[b][c]
        
    return result

V = transpose(Y)
print("5th Matrix:")
for row in V:
    print(row)

#(U) 6th matrix (multiplication of 3rd and 5th)
def matrix_multiply(A, B):
    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])

    result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result[i][j] += A[i][k] * B[k][j]

    return result

V = matrix_multiply(result, V)

print("6th Matrix:")
for row in V:
    print(row)

#(sum of elements)
sum_total = sum(sum(row) for row in result)

print("Sum of elements in 3rd matrix:")
print("=", sum_total)

#(T) 7th matrix (2x3 Zero Matrix)
T = [[0 for _ in range(3)] for _ in range(2)]

print("7th Matrix:")
for row in T:
    print(row)
```
